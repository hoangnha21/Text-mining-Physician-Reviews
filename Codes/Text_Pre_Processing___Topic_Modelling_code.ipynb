```python
def format_output(output_dict):
  polarity = "neutral"
  if(output_dict['compound']>= 0.05):
    polarity = "positive"
  elif(output_dict['compound']< 0.05):
    polarity = "negative"
  return polarity
def predict_sentiment(text):
  output_dict =  sia.polarity_scores(text)
  return format_output(output_dict)
df["vader_prediction"] = df["Review"].apply(predict_sentiment)
df
```

```python
df1['Negative_Reviews']= df1['Negative_Reviews'].astype(str)
df1['Positive_Reviews']= df1['Positive_Reviews'].astype(str)
```

```python
import nltk
nltk.download('punkt')
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
from nltk.tokenize import word_tokenize
import re

def preprocess_text(Review):
    # Convert to lowercase
    Review = Review .lower()
    # Remove special characters and punctuation
    Review = re.sub(r'[^\w\s]', '', Review)
    # Tokenize words
    words = word_tokenize(Review)
    # Remove stop words
    words = [word for word in words if word not in stopwords.words('english')]
    # Lemmatize words
    lemmatizer = WordNetLemmatizer()
    words = [lemmatizer.lemmatize(word) for word in words]
    # Rejoin words into a single string
    text = ' '.join(words)
    return text
```

```python
df1['PPNegative']  = df1['Negative_Reviews'].apply(preprocess_text)
df1['PPPositive']  = df1['Positive_Reviews'].apply(preprocess_text)
```

```python
df1
```

```python
tokens_list = []
for i in range(df1.shape[0]):
  rmv = df1.iloc[i]['Doctor Name'].lower().replace('.','').split(' ')
  str1 = df1.iloc[i]['PPNegative']
  for j in rmv:

    if str1.find(j) != -1:
      str2 = str1.replace(j, '')
      str1 = str2

  tokens_list.append(str1)
  #print(str2)
df1['PPNegative_1'] = tokens_list
```

```python
from transformers import pipeline
classifier = pipeline(
                    task = "zero-shot-classification",
                    model = "facebook/bart-large-mnli"
)
```

```python
labels = ["CME Continuing medical Education","Improper treatment or prescription","Patient Harm", "Unprofessional or immoral Conduct","failure to pay income taxes"]

context = []
for text in df["Text_of_sanction"]:
   predictions = classifier(text,
                            labels,
                            multi_label = True,
                            #hypothesis_template = template
                            )
   context.append(predictions)
```
